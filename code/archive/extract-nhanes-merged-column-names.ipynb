```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("rtweera/nhanes-training-merged", split="train") # Example for a dataset on the Hub
```

```python
column_names = dataset.column_names
```

```python
# Specify the name for your output file
output_filename = "squad_column_names.txt"

# Open the file and write each column name on a new line
with open(output_filename, 'w') as f:
    for name in column_names:
        f.write(f"{name}\n")

print(f"Column names successfully saved to {output_filename}")
```
